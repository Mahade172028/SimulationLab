```python
AX = [10]
AY = [0]

BX = [0]
BY = [10]

CX = [10]
CY = [10]

DX = [0]
DY = [0]
```

```python
import matplotlib.pyplot as plt
```

```python
SA,SB,SC,SD=3,5,7,2
A,B,C,D=0,0,0,0
```

```python
import math
```

```python
def distance(XT,XP,YT,YP):
    return math.sqrt( (XT-XP)**2 + (YT-YP)**2 )

def newPosition(XT,XP,YT,YP,S,dist):
    cos_theta = (XT - XP)/dist
    sin_theta = (YT - YP)/dist
    
    xp_new = XP + S*cos_theta
    yp_new = YP + S*sin_theta
    return xp_new,yp_new
```

```python
for t in range(0, 20):
    
    print("Time_step {} A({},{}) ,B({},{}) ,C({},{}),D({},{})".format(t,AX[t],AY[t],BX[t],BY[t],CX[t],CX[t],DX[t],DY[t]))
    dist=distance(DX[t],BX[t],DY[t],BY[t])
    
    if dist < 5:
        print("D is shoot by B at t={} dist={}".format(t,dist))
        D+=1
    xp_new,yp_new=newPosition(DX[t],BX[t],DY[t],BY[t],SB,dist)
    BX.append(xp_new)
    BY.append(yp_new)
    
    
    
    dist=distance(CX[t],AX[t],CY[t],AY[t])
    if dist < 5:
        print("C is shoot by A at t={} dist={}".format(t,dist))
        C+=1
    xp_new,yp_new=newPosition(CX[t],AX[t],CY[t],AY[t],SA,dist)
    AX.append(xp_new)
    AY.append(yp_new)
    
    
    
    dist=distance(BX[t],CX[t],BY[t],CY[t])
    if dist < 5:
        print("B is shoot by C at t={} dist={}".format(t,dist))
        B+=1
    xp_new,yp_new=newPosition(BX[t],CX[t],BY[t],CY[t],SC,dist)
    CX.append(xp_new)
    CY.append(yp_new)
    
    
    
    
    dist=distance(AX[t],DX[t],AY[t],DY[t])
    if dist < 5:
        print("A is shoot by D at t={} dist={}".format(t,dist))
        A+=1
    xp_new,yp_new=newPosition(AX[t],DX[t],AY[t],DY[t],SD,dist)
    DX.append(xp_new)
    DY.append(yp_new)
    print('\n')
        
```

```python
print("A got shoot {} times B got shoot {} times C got shoot {} times D got shoot {} times".format(A,B,C,D))
```

```python
#plt.figure(figsize=(8,8))
plt.plot(AX, AY, marker='o', label='A')
plt.plot(BX, BY, marker='o', label='B')
plt.plot(CX, CY, marker='o', label='C')
plt.plot(DX, DY, marker='o', label='D')
plt.legend()
plt.show()
```
