```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_Pi(n):
    hit=0
    hitx=[]
    hity=[]
    missx=[]
    missy=[]
    for i in range(n):
        x=np.random.rand()*5
        y=np.random.rand()*5
        if y <= np.sqrt(9-x*x):
            hit+=1
            hitx.append(x)
            hity.append(y)
        else:
            missx.append(x)
            missy.append(y)

    estimated_pi=(100*hit)/(9*n)                             # (1/4*pi*3^2)/5^2=#hit/n==100*#hit/9n
    area_of_circle=(1/4)*estimated_pi*(3^2)
    print("Estimated_pi:",estimated_pi,"Number of Hit:",hit,"Area of circle:",area_of_circle)
    
    plt.figure(figsize=(5,5))

    plt.scatter(hitx, hity, color = 'green')
    plt.scatter(missx, missy, color = 'red')
    plt.show()
    
   
    
    return (estimated_pi)
```

```python
values_of_n=[100,1000,5000,10000]
values_of_pi=[]
val_pi=0;
for n in values_of_n:
    
    
    val_pi=calc_Pi(n)
    values_of_pi.append(val_pi)
```

```python
print("Bar plot Pi vs N")
plt.bar(values_of_pi,values_of_n)
```
