```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
n =int(input())
hit=0
hitx=[]
hity=[]
missx=[]
missy=[]
for i in range(n):
    D=np.random.rand()*0.5
    theta=np.random.rand()*np.pi
    if D<=(0.5)*(math.sin(theta)):
        hit+=1
        hitx.append(theta)
        hity.append(D)
    else:
        missx.append(theta)
        missy.append(D)

estimated_pi=(2*n)/hit
print("Estimated_pi:",estimated_pi,"Number of Hit:",hit)
```

```python
plt.figure(figsize=(5,5))

plt.scatter(hitx, hity, color = 'green')
plt.scatter(missx, missy, color = 'red')
plt.show()
```
